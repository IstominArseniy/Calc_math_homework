```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy
```

Задача 1

```python
y0 = 1
eps = 0.1
N = 100
y = y0 + np.random.uniform(-eps, eps, N)
y1 = np.sum(y) / N
plt.scatter(np.arange(N), y)
plt.plot(np.linspace(0, N, N), np.ones(N) * y0, color='r')
plt.plot(np.linspace(0, N, N), np.ones(N) * y1, color='g')
print(y1)
```

Задача 2

```python
data = sns.load_dataset('anscombe')
data1x = np.array(data[data.dataset == 'I'].x)
data1y = np.array(data[data.dataset == 'I'].y)
data2x = np.array(data[data.dataset == 'II'].x)
data2y = np.array(data[data.dataset == 'II'].y)
data3x = np.array(data[data.dataset == 'III'].x)
data3y = np.array(data[data.dataset == 'III'].y)
data4x = np.array(data[data.dataset == 'IV'].x)
data4y = np.array(data[data.dataset == 'IV'].y)
fig, plots = plt.subplots(2, 2)
plots[0, 0].scatter(data1x, data1y)
plots[0, 1].scatter(data2x, data2y)
plots[1, 0].scatter(data3x, data3y)
plots[1, 1].scatter(data4x, data4y)
print("x1 dispersion is ", np.std(data1x), ", y1 dispersion is ", np.std(data1y))
print("x2 dispersion is ", np.std(data2x), ", y2 dispersion is ", np.std(data2y))
print("x3 dispersion is ", np.std(data3x), ", y3 dispersion is ", np.std(data3y))
print("x4 dispersion is ", np.std(data4x), ", y4 dispersion is ", np.std(data4y))
print("data1 pearson coef is :", np.corrcoef(data1x, data1y)[0, 1])
print("data2 pearson coef is :", np.corrcoef(data2x, data2y)[0, 1])
print("data3 pearson coef is :", np.corrcoef(data3x, data3y)[0, 1])
print("data4 pearson coef is :", np.corrcoef(data4x, data4y)[0, 1])
```

```python
l = np.array((0, 10, 20, 30.1, 40.2))
I = np.array((9.071, 8.496, 7.917, 7.358, 6.824))
result = scipy.stats.linregress(l, I)
plt.scatter(l, I)
plt.plot(l, result.intercept + result.slope * l, color='r')
print(result)
# print(result.slope, result.intercept, result.rvalue, result.stderr, result.intercept_stderr)
```
