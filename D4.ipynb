```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
N = 500
M = 500
T_steps = 5000
T = 0.005
h = 1 / max(N, M)
dt = T / T_steps
u = np.zeros((T_steps, N, M))
ux1 = np.zeros(M) + 0
ux2 = np.zeros(M) + 0
uy1 = np.zeros(N) + 0
uy2 = np.zeros(N) + 0
u[0, 0] = ux1
u[0, N-1] = ux2
u[0, :, 0] = uy1
u[0, :, M-1] = uy2
```

```python
def f(x, y):
    A = (x - 0.5) ** 2 + (y - 0.5) ** 2 < 0.02**2
    return 50 *  np.array(A)
```

```python
X, Y = np.meshgrid(np.linspace(0, 1, M, endpoint=False), np.linspace(0, 1, N, endpoint=False))
plt.imshow((f(X, Y)), cmap='Greens')
```

```python
for i in range(1, T_steps):
    u[i, 1:N-1, 1:M-1] = u[i-1, 1:N-1, 1:M-1] + dt * (f(X[1:N-1, 1:M-1], Y[1:N-1, 1:M-1]) + (u[i-1, 2:N, 1:M-1] + u[i-1, 0:N-2, 1:M-1] 
                                                                                             + u[i-1, 1:N-1, 2:M] + u[i-1, 1:N-1, 0:M-2] - 4 * u[i-1, 1:N-1, 1:M-1]) / h**2)
```
